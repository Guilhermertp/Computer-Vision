# Adaptive threshold

Adaptive threshold gives different thresholds for different regions of the same image and gives better results for images with varying ilumination.

cv2.adaptiveThreshold -> funtion fot adaptive threshold

````python
thresh=cv2.adaptiveThreshold(gray_scale,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)
cv2_imshow(thresh)
````

* Passing gray scale and the maximum value aloud is 255
* cv2.ADAPTIVE_THRESH_MEAN_C - the threshold method uses the mean of the neighbourhood area
* cv2.THRESH_BINARY - binary thresholding type where the pixels are set either to zero or 255 based on threshold
* Passing 3 and 5 , where 3 is the size of the local neighbourhood and 5 is the constant subtracted from the mean

### The ouput have some noise what doesn 't happen in the example above, but we can use filters to remove the noise

````python
# use of a gausssian filter to remove the noise of the image
image=cv2.GaussianBlur(gray_scale,(3,3),0)
cv2_imshow(image)

thresh=cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)
cv2_imshow(thresh)
````

````python
_,thresh2=cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2_imshow(thresh2) # sep the background and foreground
````



This threshold(OTSU) automaticaly calculates the optimal threshold value based on the image histogram.

*cv2.THRESH_BINARY+cv2.THRESH_OTSU - Combination of booth of threshold

# Filters for Image Processing

* What is the purpose of using filters in image processing? Filters enhance brightness, leading to improve image quality. Filters are effective at reducing noise, which is essential for restoring damaged images. Filters play a crucial role in understanding image textures and edges. Filters are used for comparing a template with each local image patch

* What is the purpose of using Gaussian filter in image processing?


<u>There are two main type of filters:</u>
* Low-pass filters
* High-pass filters

**Low-pass filters** have the purpose to allow a low frequency component to pass, it is going to smoth out/blurr the image. It is going to stop the high frequency data and allow the low frequency data, the noise will be removed. A low-pass filter (LPF) can be considered an averaging process. However, the weight and span of averaging depend on the shape and contents of the kernel.

**High-pass filters** have the purpose to allow a high frequency component to pass, it is going to sharp the image. It is going to enhance the edges. A high-pass filter (HPF) analyzes a region of an image and enhances the intensity of specific pixels by comparing their intensity with that of the surrounding pixels.


Under the category of **Low-pass filters** there are **Linear Image Filters** where the box filter and the Gaussian filter are two of the most popular linear filters
